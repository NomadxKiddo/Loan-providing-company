### 4.4 Days to Years

```python
# Lets inspect the columns in days

user_app1.loc[:, 'DAYS_BIRTH' : 'OWN_CAR_AGE']
```

```python
# Lets check for missing values
user_app1.loc[:, 'DAYS_BIRTH' : 'OWN_CAR_AGE'].isnull().sum()
```

```python
# We have missing values in 'OWN_CAR_AGE' column, Lets check weather those people dosen't really owns a car
user_app1.FLAG_OWN_CAR[user_app1.OWN_CAR_AGE.isnull()].value_counts()
```

```python
# We can see that 5 persons do own a car but the Age of the car is missing
# Lets drop such rows as they are very small in over all data set percentage

user_app1 = user_app1[~((user_app1['FLAG_OWN_CAR'] == 'Y') & (user_app1['OWN_CAR_AGE'].isnull()))]
```

```python
# Now lets check for missing values, ok we have dropped them
user_app1.FLAG_OWN_CAR[user_app1.OWN_CAR_AGE.isnull()].value_counts()
```

```python
user_app2 = user_app1.copy()
```

```python
# We can see the Data set contains the Ages, etc all in days count, lets change them to years
user_app2['YEARS_BIRTH'] = user_app1.DAYS_BIRTH.apply(lambda x : int(x / -365))
user_app2['YEARS_EMPLOYED'] = user_app1.DAYS_EMPLOYED.apply(lambda x : int(x / -365) if int(x) != 1000 else int(x))
user_app2['YEARS_REGISTRATION'] = user_app1.DAYS_REGISTRATION.apply(lambda x : int(x / -365))
user_app2['YEARS_ID_PUBLISH'] = user_app1.DAYS_ID_PUBLISH.apply(lambda x : int(x / -365))
user_app2['YEARS_OWN_CAR_AGE'] = user_app1.OWN_CAR_AGE.apply(lambda x : int(x / -365) if x > 0 else x)
```
